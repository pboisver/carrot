## Step 1: Initialize the Module
Now we import our custom `diet_safety` module and instantiate the manager class. It will automatically read our `animals.csv` dataset file.

```python
from carrot.diet_safety import DietSafetyManager

# Create an instance of our module helper
manager = DietSafetyManager(csv_path='src/carrot/animals.csv')
print("Module loaded and data populated successfully!")
```

## Step 2: Individual Animal Checks
Let's check the dynamic relationship status of a few individual animals.

```python
test_animals = ["Chicken", "Lion", "Grizzly Bear", "Alien"]

for animal in test_animals:
    status = manager.get_relationship(animal)
    print(f"{animal} -> {status}")
```

## Step 3: Categorized Listings
We can extract lists of specific threats and opportunities from our data layer using the module's categorization filters.

```python
print("Animals that ONLY eat me (Avoid):\n", manager.list_by_category('animal'))
print("\nAnimals that ONLY I eat (Safe Hunt):\n", manager.list_by_category('me'))
print("\nAnimals where we eat each other (Mutual Danger):\n", manager.list_by_category('both'))
```
